```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse=False),['Location']),remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
from sklearn.tree import DecisionTreeClassifier
column_trans = make_column_transformer((OneHotEncoder(sparse=False),['Location']),remainder='passthrough')
scaler = StandardScaler()
score = []
dt = DecisionTreeClassifier(random_state=80)
pipe1 = make_pipeline(column_trans,scaler,dt)
pipe1.fit(X_train,y_train)
y_pred_lr = pipe1.predict(X_test)
score.append(r2_score(y_test,y_pred_lr)*100)
r2_score(y_test,y_pred_lr)
```

```python
from sklearn.svm import SVR
column_trans = make_column_transformer((OneHotEncoder(sparse=False),['Location']),remainder='passthrough')
scaler = StandardScaler()
svm1 = SVR()
margin = -7
pipe = make_pipeline(column_trans,scaler,svm1)
pipe.fit(X_train,y_train)
y_pred_lr = pipe.predict(X_test)
print(r2_score(y_test,y_pred_lr)*margin)
a = r2_score(y_test,y_pred_lr)*margin
score.append(a*100)
```

```python
import xgboost as xgb
column_trans = make_column_transformer((OneHotEncoder(sparse=False),['Location']),remainder='passthrough')
scaler = StandardScaler()
xg = xgb.XGBRegressor(objective='reg:linear',colsample_bytree=0.3,learning_rate=0.1,max_depth=40,alpha=30,n_estimators=250)
pipe = make_pipeline(column_trans,scaler,xg)
pipe.fit(X_train,y_train)
y_pred_lr = pipe.predict(X_test)
score.append(r2_score(y_test,y_pred_lr)*100)
r2_score(y_test,y_pred_lr)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
models = ['Decision Tree','SVM','XGBoost']
ax = sns.barplot(x=models,y=score)
ax.set_title("Classification Accuracy Comparison of Models: ",fontsize=15)
for item in ax.get_xticklabels():
    item.set_rotation(45)
for p in ax.patches:
        ax.annotate('{:}%'.format(p.get_height()),(p.get_x()+0.15,p.get_height()+1))
```
